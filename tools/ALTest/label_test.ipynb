```python
import mmcv
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_name = '/media/alan/新加卷1/Master/日常工作/帮助身边人/SonLiu/Detection/codes/data/Liu/images/train/wired_001.jpg'
label_file_name = '/media/alan/新加卷1/Master/日常工作/帮助身边人/SonLiu/Detection/codes/data/Liu/labels/train/wired_001.txt'
regions = np.loadtxt(label_file_name)
img = mmcv.imread(file_name)
height, width = img.shape[:2]

plt.imshow(img)
plt.show()
```

```python
import cv2
region = regions[0]
x, y, w, h = region[1], region[2], region[3], region[4]

x1 = (x - w / 2) * width
y1 = (y - h / 2) * height
x2 = (x + w / 2) * width
y2 = (y + h / 2) * height
img = cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
plt.imshow(img)
```
